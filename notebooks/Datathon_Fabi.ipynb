##3.9 Modelo Preditivo de Risco

```python
df['Risco'] = np.where(df['IAN'] < 5.0, 1, 0)
df['Risco'].value_counts()
```

#4. Modelo Preditivo

## 4.1 Selação de features

```python
features = ['IDA','IEG','IAA','IPS','IPV']
X = df[features]
y = df['Risco']

# Tratamento de nulos
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

##4.2 Modelo Ramdon Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, model.predict_proba(X_test)[:,1]))
```

#5. Análise das Variáveis

##5.1 Importancia das Variáveis

```python
importances = pd.Series(model.feature_importances_, index=features)
importances.sort_values().plot(kind='barh')
plt.title("Importância das Variáveis")
plt.show()
```

##5.2 Efetividade do Programa por Fase

```python
importances = pd.Series(model.feature_importances_, index=features)
importances.sort_values().plot(kind='barh')
plt.title("Importância das Variáveis")
plt.show()
```
